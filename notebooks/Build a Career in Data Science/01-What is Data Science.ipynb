# Chapter 1: What is Data Science?

At a good company, data scientists enjoy a lot of autonomy and are constantly learning new things. They use their skills to solve significant problems with data. 

But not all data science jobs are perfect. Both companies and job seekers can have unrealistic expectations. Companies new to data science may think that one person can solve all their problems with data, for example. On the other side, job seekers may think there will never be a dull moment in their new career. They may expect that stakeholders will follow their recommendations rutinely, that data engineers can fix any data quality issues immediately, and that they'll get the faster computing resources available to implement their models. 

In reality, data scientists spend a lot of time cleaning and preparing data, as well as managing the expectations and priorities of other teams. Projects won't always work out. Data scientists may notice lots of satistical or technical mistakes in legacy analyses that have real consequences, but no one is interested. 

***So what is data science?***

Data science is the practice of using data to try to understand and solve real-world problems. This is not a new concept; people have been analyzing data since the invention of the zero. In the past 1-2 decades, however, we have gained access to exponentially more data than ever existed before. The advent of computers has assisted in the generation of all that data. 

When many people start to look into data science, one challenge they may face is being overwhelmed by the amount of things they need to learn, such as coding, statistics, machine learning, and domain knowledge.  In addition, they need to learn business smkills such as effectively communicating results to audiences ranging from other data scientists to the CEO. How can you possibly learn all of these skills? Which ones should you start with? What are the basics?

This book offers the opinion that data science is a venn diagram of three main components:
1) Math and Stats
2) Programming and Databases
3) Domain Knowledge

They also make the point that skills are attached to the intersections of these components. The intersections between:
* Domain Knowledge and Math and Stats is ***Decision Science***
* Domain Knowledge and Programming and Databases is ***Analytics***
* Programming and Databases and Math and Stats is ***Machine Learning***

All 3 skills are fundamental to data science, however each data scientist may posses a different extent of skills.

***Math and Stats***

At the basic level, math + stats is *data literacy*. You must know:
* that techniques exist
* how to apply techniques - with code and the technicality behind them
* how to choose which techniques to try

***Programming and Databases***

This skill refers to your ability to pull data from the company databases and write clean, efficient, maintainable code. These skills are similar to what a software developer has to know, except the data scientists have to write code that does open-ended analysis rather than produces a predefined output. Each company's data stack is unique, but broadly you will need to know how to get data from a database and how to clean, manipulate, summarize, visualize, and share data.

You will need to know:
* Programming languages (R or Python)
* SQL
* Version control (Git/GitHub)

***Domain Knowledge***

A core skill in data science is translating a business situation into a data question, finding the data answer, and delivering it as a business answer. Bussiness understanding and domain knowledge help you know what questions to ask next. 

Another part of business understanding is developing general business skills, such as being able to tailor your presentations and reports to different audiences. 

Finally, as you become more senior, part of your job is to identify where the business could benefit from data science. 


### Different Types of Data Science Jobs

You can mix and match the three core skills of data science into several jobs - all of which have some justification for having the title data scientist. When looking for data science jobs, you should pay less attention to the job titles and more to the job description. 

***Analytics***

An analyst takes data and puts it in front of the right people. This work involves a lot of data cleaning and preparation, but generally less work to interpret the data. Although you should be able to spot and fix data quality issues, the primary person who makes decisions with this data is the business partner. 

Short rule: An analyst creates dashboards and reports that deliver data.

***Machine Learning***

A machine learning engineer develops machine learning models and puts them into production, where they run continously. The MLE spends  less time on things like creating visualizations and dashboards, and more time doing the programming work of data science. 

In many ways, this roles is closer to a software developer than other data science roles are. Your code must be performant, tested, and written so that other people will be able to wrok with it. 

Short rule: a machine learning engineer creates models that get run continously.

***Decision Science***

A decision scientist turns a companies raw data into information that helps the company make decisions. This work relies on having deep understanding of different mathematical and statistical methods and familiarity with business decisions. Although a decision scientist does plenty of programming, their work generally gets run only once to make a particular analysis. A decision scientist often users their knowledge of statistics to help the company make decisions under uncertainty. 

Short rule: a decision scientist creates analyses that produce recommendations.

### Related Jobs

***Business Intelligence Analyst***

This work is similar to that of an analyst, however, these rules generally use less statistical and programming skills. Their tool of choice may be Excel. 

***Data Engineer***

A data engineer focuses on keeping data maintained in databses and ensuring that people can get the data they need. They don't run reports, make analyses, or develop models; instead they keep data neatly stored and formatted in well-structured databases so that other people can do those things. 

***Research Scientist***

A research scientist develops and implements new tools, algorithms, and methodologies, often to be used by other data scientists within the company. These types of positions almost always require PhDs.
