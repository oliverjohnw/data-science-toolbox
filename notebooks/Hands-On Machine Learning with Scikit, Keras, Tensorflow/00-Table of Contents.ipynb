### Preface

The book relies on the following 3 Python packages:
* **Scikit-Learn**: ML algorithms.
* **Tensorflow**: More complex the scikit, used for distributed numerical computation. Can train + run large Neural Networks across GPUs.
* **Keras**: High-level deep learning API. Built on Tensorflow and relies on it for all the intensive computations.

All code examples can be found at https://github.com/ageron/handson-ml3.
